```python
##questão A:
df_delay_min = df_new[df_new["dep_delay"]>5]
df_delay_min
```

```python
##questão B:
df_group_dest = df_delay_min.groupby(["dest"]).count().reset_index()
#df_group_dest
df_dest_sort = df_group_dest.sort_values(by=['dep_delay'], ascending=False)
df_dest_sort.tail(5)
```

ETAPA 3

Quais são as velocidades médias dos voos nos 5 destinos que mais ocorrem 
atrasos.

a. Filtrar a tabela com os registrados que mais atrasam

b. Calcular a velocidade média

    i. Criar uma coluna com os valores dessas velocidades médias usando as colunas de ['distance'] e ['air_time']

    ii. Agrupar a tabela com os piores destinos

```python
##questão A:
df_dest_list = df_dest_sort.head(5).dest
df_topdelay_dest = df_new.query('dest in @df_dest_list')
df_topdelay_dest
```

```python
##questão B:
df_topdelay_dest['velocidade_media'] = df_topdelay_dest['distance'] / (df_topdelay_dest['air_time']/60)
df_topdelay_dest
```

```python
##questão B:
df_dst_group = df_topdelay_dest.groupby(["dest"]).mean().reset_index()
df_dst_group
```

```python
#sns.countplot(data=df_dst_group,hue='dest',y='velocidade_media')
px.histogram(df_dst_group,y='dest',x='velocidade_media')
```
